```python
import os
from langchain_community.document_loaders import PyPDFLoader

# Corrige o caminho para incluir a pasta chatbot-top-contratacoes-agu
pdf_path = os.path.join(os.getcwd(), 'chatbot-top-contratacoes-agu', 'inputs', 'Instrumento_AGU.pdf')

loader = PyPDFLoader(pdf_path)
documents = loader.load()

print(f"Total de páginas carregadas: {len(documents)}")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Dividindo o texto em pedaços menores
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,    # cada pedaço terá no máximo 800 caracteres
    chunk_overlap=50   # e 50 caracteres de sobreposição entre eles
)

texts = text_splitter.split_documents(documents)

print(f"Total de pedaços de texto gerados: {len(texts)}")
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma

# Criar o objeto de embeddings localmente
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# Criar a base vetorial usando ChromaDB
db = Chroma.from_documents(texts, embeddings)

print("Embeddings criados localmente e armazenados com sucesso!")
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.chains import RetrievalQA

# Aqui não precisa mais da OpenAI, usaremos uma resposta fictícia só para teste (depois melhoramos)
from langchain.llms.fake import FakeListLLM

# LLM Fake só para simular respostas (até depois você plugar um modelo real)
llm = FakeListLLM(responses=["Essa resposta foi gerada a partir do seu documento!"])

# Conectar a busca vetorial ao modelo
qa = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=db.as_retriever()
)

# Função para perguntar
def perguntar(questao):
    resposta = qa.run(questao)
    return resposta

# Exemplo:
pergunta = "O que é o Instrumento de Padronização de Procedimentos da AGU?"
resposta = perguntar(pergunta)
print(resposta)
```
